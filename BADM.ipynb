```python
#Importing Pandas to read dataset as a dataframe
import pandas as pd
import numpy as np
```

```python
np.random.seed(25)
```

```python
#Importing the dataset
df = pd.read_csv('auto-mpg.csv')
```

```python
#Dataset info
df.info()
```

```python
#Reading 10 Random samples from the Dataset
df.sample(10)
```

### Dropping the unrequired columns "car name" & "origin"

```python
#Dropping the columns 'car name' and 'origin'
columns = ['car name', 'origin']
df.drop(columns, axis = 1,inplace = True )
```

```python
#Printing random 10 samples from the dataframe
df.sample(10)
```

```python
#Checking rows with ? in horsepower
df[df['horsepower'] == '?']
```

### As we can see there are some "?" in Horsepower column we will be dropping these rows

```python
# Creating a new dataframe where horsepower is not equal to "?"
df_new = df[df['horsepower'] != '?'] 

# Print the new dataframe shape
df_new.shape 
```

### We have a total of 392 records (rows) and 7 properties (columns)

```python
df_new['horsepower'] = df_new['horsepower'].astype('int64')
```

```python
#Checking for the type of horsepower
df_new.info()
```
